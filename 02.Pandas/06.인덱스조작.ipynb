### 2.6 인덱스 조작

```python
import numpy as np
import pandas as pd
```

- 인덱스 설정 및 제거

```python
np.random.seed(0)
df1 = pd.DataFrame(np.vstack([list('ABCDE'),
                              np.round(np.random.rand(3, 5), 2)]).T,    #T : transpose(방향을 돌림)
                   columns=["C1", "C2", "C3", "C4"])
df1
```

```python
df2 = df1.set_index('C1')   #C1을 인덱스명으로 지칭
df2
```

```python
print(df1.index)
print(df2.index)
```

```python
df2.reset_index()
```

```python
df1.to_csv('data/sample11.csv', index=False)
pd.read_csv('data/sample11.csv')
```

```python
df2.to_csv('data/sample12.csv')
pd.read_csv('data/sample12.csv', index_col='C1')
```

- 다중 인덱스

```python
np.random.seed(0)
df3 = pd.DataFrame(np.round(np.random.randn(5, 4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C1", "C2", "C1", "C2"]])
df3
```

```python
# 0.41
df3['A','C2'][2]
```

```python
df3['A']
```

```python
df3.columns.names = ['Cidx1', 'Cidx2']
df3
```

```python
np.random.seed(0)
df4 = pd.DataFrame(np.round(np.random.randn(6, 4), 2),
                   columns=[["A", "A", "B", "B"],
                            ["C", "D", "C", "D"]],
                   index=[["M", "M", "M", "F", "F", "F"],
                          ["id_" + str(i + 1) for i in range(3)] * 2])
df4.columns.names = ["Cidx1", "Cidx2"]
df4.index.names = ["Ridx1", "Ridx2"]
df4
```

```python
# 0.41
df4['A','D']['M','id_3']
```

```python
# iloc 인덱서를 사용하면 다중 인덱스 무시하고 인덱싱하면 됨
df4.iloc[2,1]
```

```python
# 연습 문제 4.5.1
np.random.seed(0)
df_score1 = pd.DataFrame(np.random.randint(10,101).reshape(5,3),columns=["국어", "영어", "수학"])
df_score1
```

- 1.학생 이름을 나타내는 열을 포함시키지 않고 데이터프레임 df_score1 을 생성한 후, \
- df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다.
- reset_index 명령으로 이 인덱스 열을 명령으로 일반 데이터열로 바꾸여 데이터프레임 df_score2을 
- 만든다.

```python
df_score1 = pd.DataFrame(
    np.random.randint(10, 101, 15).reshape(5, 3),
    columns=['국어','영어','수학']
)
df_score1
```

```python
df_score1['학생이름'] = ['김철수', '박명수', '노홍철', '길', '유재석']
df_score1
```

```python
df_score2 = df_score1.reset_index()
df_score2 
```

- 학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 df_score2에 set_index 명령을 
- 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다.

```python
df_score2.set_index("학생이름")
```
