- 인덱스 설정 및 제거

```python
np.random.seed(0)
df1 = pd.DataFrame(np.vstack([list('ABCDE'),
                              np.round(np.random.rand(3, 5), 2)]).T,    #T : transpose(방향을 돌림)
                   columns=["C1", "C2", "C3", "C4"])
df1
```

```python
df2.reset_index()
```

```python
# 연습 문제 4.5.1
np.random.seed(0)
df_score1 = pd.DataFrame(np.random.randint(10,101).reshape(5,3),columns=["국어", "영어", "수학"])
df_score1
```

- 1.학생 이름을 나타내는 열을 포함시키지 않고 데이터프레임 df_score1 을 생성한 후, \
- df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다.
- reset_index 명령으로 이 인덱스 열을 명령으로 일반 데이터열로 바꾸여 데이터프레임 df_score2을 
- 만든다.

```python
df_score1 = pd.DataFrame(
    np.random.randint(10, 101, 15).reshape(5, 3),
    columns=['국어','영어','수학']
)
df_score1
```

```python
df_score1['학생이름'] = ['김철수', '박명수', '노홍철', '길', '유재석']
df_score1
```

```python
df_score2 = df_score1.reset_index()
df_score2 
```

- 학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 df_score2에 set_index 명령을 
- 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다.

```python
df_score2.set_index("학생이름")
```

```python
# 선생님 답
df_score1 = pd.DataFrame(np.random.randint(60, 101, size=(5, 3)), 
                         columns=['국어', '영어', '수학'])
df_score1
```

```python
df_score1.index = ['제임스','마리아','파이썬','김자바','최디비']
df_score1.index.name = '이름'       #이름이라는 인덱스를 만들어 학생이름을 넣는다.
df_score1
```

```python
df_score2 = df_score1.reset_index()     #이름인덱스를 df_score2에 새로 열로 넣는다
df_score2
```

```python
df_score2.set_index('이름', inplace=True)   #다시 학생이름 열을 인덱스로 변경
df_score2
```

```python
# CCTV 문제 
cctv = pd.read_csv('data/서울시 자치구 연도별 방범용 CCTV 운영 현황_230630기준.csv', encoding='EUC-KR', index_col='구분')
del cctv['Unnamed: 0']
cctv.head()
```

```python
cctv.dtypes
```

```python
for col in cctv.columns:
    print(col, type(cctv[col][0]))
```

```python
# 선생님 답
for col in cctv.columns:
    if col in ['2021년', '2022년']:
        continue
    cctv[col] = cctv[col].str.replace(',','').astype(int) #replace할때 str.replace로 써야함!
```

```python
cctv['계'] = cctv.sum(axis=1)
cctv.head()
```

```python
# 천단위 구분기호 옵션을 설정하면 쉽게 읽을 수 있음
cctv = pd.read_csv('data/서울시 자치구 연도별 방범용 CCTV 운영 현황_230630기준.csv', encoding='EUC-KR', index_col='구분', thousands=',')
del cctv['Unnamed: 0']
cctv.head()
```

```python
cctv['계'] = cctv.sum(axis=1)
cctv.tail()
```
